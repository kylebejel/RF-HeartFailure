```python
import pandas as pd
import numpy as np
```

```python
Heart_failure = pd.read_csv('../input/heart-failure-prediction/heart.csv')
Heart_failure.head()
```

```python
Heart_failure['Sex'] = Heart_failure['Sex'].map({'M':1,'F':0})
Heart_failure['ExerciseAngina'] = Heart_failure['ExerciseAngina'].map({'Y':1,'N':0})
```

```python
status = pd.get_dummies(Heart_failure['ChestPainType'], drop_first = True)
Heart_failure = pd.concat([Heart_failure, status], axis = 1)
Heart_failure.drop(['ChestPainType'], axis = 1, inplace = True)

status = pd.get_dummies(Heart_failure['RestingECG'], drop_first = True)
Heart_failure = pd.concat([Heart_failure, status], axis = 1)
Heart_failure.drop(['RestingECG'], axis = 1, inplace = True)

status = pd.get_dummies(Heart_failure['ST_Slope'], drop_first = True)
Heart_failure = pd.concat([Heart_failure, status], axis = 1)
Heart_failure.drop(['ST_Slope'], axis = 1, inplace = True)
```

```python
X = Heart_failure.drop('HeartDisease',axis=1)
y = Heart_failure['HeartDisease']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7,random_state=100)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
Model = RandomForestClassifier(random_state=100,n_jobs=-1)
Model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,recall_score,plot_roc_curve
```

```python
plot_roc_curve(Model,X_test,y_test)
y_pred = Model.predict(X_test)

print(accuracy_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
```
